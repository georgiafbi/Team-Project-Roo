```python
import pandas as pd
import matplotlib.pyplot as plt
# Dependencies
import requests
from pprint import pprint
import numpy as np
import gmaps
import json
import os
from scipy.stats import linregress
from datetime import datetime
import time

# Google developer API key
#from config import g_key

stations_df = pd.read_csv("csv_files/weather_stations_codes.csv", sep=";")

#renames columns to english
stations_df=stations_df.rename(columns={"Nome": "Name","Código": "Code", "Status da Operação":"Operation Status", "Ínicio da Operação": "Start of Operation"})

#drops unwanted columns "Operation Status","Start of Operation","Name"
stations_ndf=stations_df.drop(["Operation Status","Start of Operation"], axis=1)

#Splits station names to get state abbreviation
stations_ndf[["Station", "State"]]=stations_ndf.Name.str.split("- ",expand=True)

#locates only states where deforestation is occuring with lat and lng. 
states=['AC','AP',"AM","MA","MT","PA","RR","TO"]
station_lat_lng=stations_ndf.loc[stations_ndf["State"].isin(states)]
station_lat_lng=station_lat_lng.drop(["Name", "Altitude"], axis=1)
station_lat_lng.to_csv("useable_csv/lat_lng_station.csv")
station_lat_lng
```

```python
#Drops "Latitude", "Longitude", "Altitude"
stations_ndf=stations_ndf.drop(["Name", "Latitude", "Longitude", "Altitude"],axis=1)

#renaming "Station": "Station Name", "Code":"Station"
stations_ndf=stations_ndf.rename(columns={"Station": "Station Name", "Code":"Station"})
#saves station with dropped lat and long to altitude
stations_ndf.to_csv("useable_csv/station_and_state.csv")

stations_ndf
```
